```python
# =========================
# COMPLETE GPU LIGHTGBM - FIXED DTYPE ISSUE
# =========================
import time
import pandas as pd
import numpy as np
import joblib
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# =========================
# PATHS
# =========================
data_path = '/content/drive/MyDrive/JANI/Loan_default.csv'
save_path = '/content/drive/MyDrive/JANI/fast_gpu_model.pkl'

# =========================
# LOAD & CLEAN DATA
# =========================
print("Loading data...")
df = pd.read_csv(data_path)
print(f"Dataset: {df.shape}")

# Drop ID column, separate target
X = df.drop(columns=["Default", "LoanID"])  # Drop LoanID too
y = df["Default"]

# Convert ALL object columns to category dtype (FIXES THE ERROR)
object_cols = X.select_dtypes(include=["object"]).columns
for col in object_cols:
    X[col] = X[col].astype("category")

print(f"Categorical features converted: {len(object_cols)}")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# =========================
# GPU OPTIMIZED PARAMS
# =========================
scale_pos_weight = (y_train==0).sum() / (y_train==1).sum()
print(f"Class imbalance ratio: {scale_pos_weight:.1f}")

params = {
    'device': 'gpu',
    'gpu_platform_id': 0,
    'gpu_device_id': 0,
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'max_depth': -1,
    'min_child_samples': 20,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'scale_pos_weight': scale_pos_weight,
    'max_bin': 63,  # GPU optimized
    'random_state': 42,
    'verbose': -1,
    'force_col_wise': True,
    'num_threads': 2  # Limit CPU threads for GPU
}

# =========================
# TRAIN ON GPU
# =========================
print("🚀 GPU Training started...")
start = time.perf_counter()

# LightGBM handles category dtype automatically now
train_data = lgb.Dataset(X_train, label=y_train, free_raw_data=False)
valid_data = lgb.Dataset(X_test, label=y_test, reference=train_data, free_raw_data=False)

model = lgb.train(
    params,
    train_data,
    valid_sets=[valid_data],
    num_boost_round=2000,
    callbacks=[lgb.early_stopping(100), lgb.log_evaluation(50)]
)

end = time.perf_counter()
print(f"✅ Training complete: {end-start:.1f}s")

# =========================
# EVALUATE & SAVE
# =========================
y_prob = model.predict(X_test)
y_pred = (y_prob > 0.5).astype(int)

auc_score = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC Score: {auc_score:.4f}")
print(f"Predictions shape: {y_pred.shape}")

# Save model
joblib.dump(model, save_path)
print(f"💾 Model saved: {save_path}")

# Feature importance
importance = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importance()
}).sort_values('importance', ascending=False)

print("\nTop 10 Features:")
print(importance.head(10))
```

```python
# =========================
# ULTRA-ACCURATE GPU LIGHTGBM - GPU ERROR FIXED
# =========================
import time
import pandas as pd
import numpy as np
import joblib
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.preprocessing import LabelEncoder

# =========================
# PATHS
# =========================
data_path = '/content/drive/MyDrive/JANI/Loan_default.csv'
save_path = '/content/drive/MyDrive/JANI/perfect_gpu_model.pkl'

# =========================
# LOAD & ENGINEERED PREPROCESSING
# =========================
print("Loading data...")
df = pd.read_csv(data_path)
print(f"Dataset: {df.shape}")

X = df.drop(columns=["Default", "LoanID"])
y = df["Default"]

# NUMERIC ENCODING ONLY (GPU SAFE)
print("Encoding categorical features...")
label_encoders = {}
for col in X.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le

print(f"Encoded {len(label_encoders)} features")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# =========================
# GPU-SAFE HIGH-ACCURACY PARAMS
# =========================
scale_pos_weight = (y_train==0).sum() / (y_train==1).sum()
print(f"Class imbalance: {scale_pos_weight:.1f}")

params = {
    # GPU Settings (FIXED)
    'device': 'gpu',
    'gpu_platform_id': 0,
    'gpu_device_id': 0,

    # Core
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',

    # HIGH ACCURACY
    'learning_rate': 0.02,
    'num_leaves': 50,
    'max_depth': 7,
    'min_data_in_leaf': 100,
    'feature_fraction': 0.85,
    'bagging_fraction': 0.85,
    'bagging_freq': 5,
    'scale_pos_weight': scale_pos_weight,

    # GPU OPTIMIZED (REMOVED PROBLEMATIC PARAMS)
    'max_bin': 63,           # SAFE for GPU
    'num_threads': 2,

    # Regularization
    'lambda_l1': 0.05,
    'lambda_l2': 0.05,

    'seed': 42,
    'verbose': -1
}

# =========================
# BULLETPROOF GPU TRAINING
# =========================
print("🚀 Ultra-Accurate GPU Training...")
start = time.perf_counter()

train_data = lgb.Dataset(X_train, label=y_train, free_raw_data=False)
valid_data = lgb.Dataset(X_test, label=y_test, reference=train_data, free_raw_data=False)

model = lgb.train(
    params,
    train_data,
    valid_sets=[valid_data],
    num_boost_round=4000,
    callbacks=[
        lgb.early_stopping(150),
        lgb.log_evaluation(100)
    ]
)

end = time.perf_counter()
print(f"✅ Training complete: {end-start:.1f}s")

# =========================
# FULL EVALUATION
# =========================
y_prob = model.predict(X_test)
y_pred = (y_prob > 0.5).astype(int)

auc = roc_auc_score(y_test, y_prob)
print(f"\n🎯 ROC-AUC: {auc:.4f}")
print("\n📊 Classification Report:")
print(classification_report(y_test, y_pred))

# Feature importance
importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importance(importance_type='gain')
}).sort_values('importance', ascending=False)

print("\n🔥 Top 10 Features:")
print(importance_df.head(10))

# =========================
# SAVE COMPLETE PIPELINE
# =========================
joblib.dump({
    'model': model,
    'encoders': label_encoders,
    'features': X.columns.tolist(),
    'params': params
}, save_path)

print(f"\n💾 SAVED: {save_path}")
print(f"⏱️  TOTAL TIME: {time.perf_counter() - start:.1f}s")
```
